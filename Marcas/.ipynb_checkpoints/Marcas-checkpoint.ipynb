```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import numpy as np
```

```python
url = 'https://mifarma.es/marcas/'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content)
```

```python
marcas = soup.find_all ('ul', class_='bare-list')
```

```python
marcas
```

```python
listado_marcas = [elemento.text for elemento in marcas]
listado_marcas
```

```python
df_marcas = pd.DataFrame(listado_marcas)
df_marcas
```

```python
df_marcas.to_csv("marcas.csv")
```
