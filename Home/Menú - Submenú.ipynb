### Men√∫

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import numpy as np
```

```python
url = 'https://mifarma.es'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content)
```

```python
menu = soup.find_all(class_="level0 has-children")
menu
```

```python
len(menu)
```

```python
df_menu = pd.DataFrame(menu)
df_menu
```

```python
df_menu.to_csv("men√∫.csv")
```
